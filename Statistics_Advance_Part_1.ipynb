In essence, the CDF accumulates the probabilities from the PDF up to a certain point.

2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of the discrete uniform distribution
lower_bound = 1
upper_bound = 10

# Number of random samples to generate
num_samples = 1000

# Generate random samples from a discrete uniform distribution
random_samples = np.random.randint(lower_bound, upper_bound + 1, size=num_samples)

# Calculate the frequency of each value
unique_values, counts = np.unique(random_samples, return_counts=True)

# Calculate the probability of each value (PMF)
probabilities = counts / num_samples

# Plot the Probability Mass Function (PMF)
plt.figure(figsize=(8, 6))
plt.bar(unique_values, probabilities, align='center', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Probability Mass Function of a Discrete Uniform Distribution')
plt.xticks(unique_values)
plt.grid(axis='y', linestyle='--')
plt.show()
```

3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
def bernoulli_pdf(p, x):
  """
  Calculates the Probability Distribution Function (PDF) of a Bernoulli distribution.

  Args:
    p: The probability of success (a float between 0 and 1).
    x: The outcome (either 0 for failure or 1 for success).

  Returns:
    The probability of the given outcome.
  """
  if x == 1:
    return p
  elif x == 0:
    return 1 - p
  else:
    raise ValueError("Outcome x must be either 0 or 1")

# Example usage:
probability_of_success = 0.7
outcome_is_success = 1
outcome_is_failure = 0

pdf_success = bernoulli_pdf(probability_of_success, outcome_is_success)
pdf_failure = bernoulli_pdf(probability_of_success, outcome_is_failure)

print(f"The PDF for outcome 1 (success) with p={probability_of_success} is: {pdf_success}")
print(f"The PDF for outcome 0 (failure) with p={probability_of_success} is: {pdf_failure}")
```

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of the discrete uniform distribution
lower_bound = 1
upper_bound = 10

# Generate possible values in the distribution
values = np.arange(lower_bound, upper_bound + 1)

# Calculate the probability of each value (for a discrete uniform distribution, it's equal for all values)
probability_per_value = 1 / len(values)

# Calculate the Cumulative Distribution Function (CDF)
# The CDF at a given value is the sum of probabilities of all values less than or equal to it.
cdf_values = [np.sum(probability_per_value for val in values if val <= x) for x in values]

# Plot the CDF
plt.figure(figsize=(8, 6))
plt.step(values, cdf_values, where='post')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF) of a Discrete Uniform Distribution')
plt.xticks(values)
plt.grid(axis='y', linestyle='--')
plt.show()
```

9. Write a Python function to calculate Z-scores from a dataset and plot them>

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
  """
  Calculates Z-scores for a dataset and plots their distribution.

  Args:
    data: A list or NumPy array of numerical data.

  Returns:
    A NumPy array of calculated Z-scores.
  """
  # Calculate the mean and standard deviation of the dataset
  mean = np.mean(data)
  std_dev = np.std(data)

  # Calculate Z-scores for each data point
  z_scores = (data - mean) / std_dev

  # Plot the histogram of Z-scores
  plt.figure(figsize=(8, 6))
  plt.hist(z_scores, bins=30, density=True, alpha=0.7, edgecolor='black')
  plt.xlabel('Z-score')
  plt.ylabel('Probability Density')
  plt.title('Distribution of Z-scores')
  plt.grid(axis='y', linestyle='--')
  plt.show()

  return z_scores

# Example usage with a sample dataset
sample_data = np.array([10, 12, 15, 11, 13, 16, 14, 12, 10, 15, 25])
z_scores = calculate_and_plot_zscores(sample_data)

print("Calculated Z-scores:")
print(z_scores)
```

11. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem

  The Central Limit Theorem (CLT) can be demonstrated by simulating multiple samples from a distribution, calculating the mean of each sample, and observing that the distribution of these sample means tends towards a normal distribution, regardless of the original distribution's shape, provided the sample size is sufficiently large.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
num_samples = 10000
sample_size = 30
original_distribution_mean = 5
original_distribution_std = 2

# Create a normal distribution (as requested by the user)
original_distribution = np.random.normal(loc=original_distribution_mean, scale=original_distribution_std, size=num_samples)

# Store sample means
sample_means = []

# Simulate multiple samples and calculate their means
for _ in range(num_samples):
    # Generate a sample from the original distribution
    sample = np.random.choice(original_distribution, size=sample_size, replace=True)
    # Calculate the mean of the sample
    sample_mean = np.mean(sample)
    # Store the sample mean
    sample_means.append(sample_mean)

# Plot the original distribution and the distribution of sample means
plt.figure(figsize=(12, 5))

# Original distribution
plt.subplot(1, 2, 1)
plt.hist(original_distribution, bins=50, density=True, alpha=0.6, color='blue', label='Original Distribution (Normal)')
plt.title('Original Distribution (Normal)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

# Distribution of sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=50, density=True, alpha=0.6, color='red', label='Sample Means')
plt.title('Distribution of Sample Means (CLT)')
plt.xlabel('Mean Value')
plt.ylabel('Density')

# Overlay a normal distribution for comparison
x = np.linspace(min(sample_means), max(sample_means), 100)
plt.plot(x, 1/(np.std(sample_means) * np.sqrt(2 * np.pi)) * np.exp( - (x - np.mean(sample_means))**2 / (2 * np.std(sample_means)**2) ),
         linewidth=2, color='green', label='Normal Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```

16. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():
  """
  Calculates and plots the Probability Density Function (PDF)
  of the standard normal distribution (mean=0, std=1).
  """
  # Define the range of x values
  x = np.linspace(-4, 4, 100) # Typically plot from -4 to 4 standard deviations

  # Calculate the PDF of the standard normal distribution
  pdf_values = norm.pdf(x, loc=0, scale=1)

  # Plot the PDF
  plt.figure(figsize=(8, 6))
  plt.plot(x, pdf_values, linewidth=2, color='blue')
  plt.title('Standard Normal Distribution (Mean=0, Std Dev=1)')
  plt.xlabel('Z-score')
  plt.ylabel('Probability Density')
  plt.grid(True, linestyle='--', alpha=0.6)
  plt.show()

# Call the function to plot the standard normal distribution
plot_standard_normal_distribution()
```

18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
  """
  Calculates the Z-score for a given data point.

  Args:
    data_point: The individual data point.
    mean: The mean of the dataset.
    std_dev: The standard deviation of the dataset.

  Returns:
    The calculated Z-score.
  """
  z_score = (data_point - mean) / std_dev
  return z_score

# Example usage:
data_point = 18
mean = 15
std_dev = 3

z_score = calculate_z_score(data_point, mean, std_dev)

print(f"The data point {data_point} has a Z-score of: {z_score:.2f}")

# To compare it to a standard normal distribution, we can visualize it
# Plot the standard normal distribution
x = np.linspace(-4, 4, 100)
plt.figure(figsize=(8, 6))
plt.plot(x, norm.pdf(x, 0, 1), label='Standard Normal Distribution (PDF)')

# Mark the calculated Z-score on the plot
plt.axvline(z_score, color='red', linestyle='dashed', linewidth=2, label=f'Z-score = {z_score:.2f}')

plt.title('Comparison of Z-score to Standard Normal Distribution')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# You can also find the probability of getting a value less than or equal to this Z-score
cumulative_probability = norm.cdf(z_score)
print(f"The cumulative probability up to a Z-score of {z_score:.2f} is: {cumulative_probability:.4f}")
```

23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution

```python
import numpy as np
from scipy.stats import poisson

# Define the parameter for the Poisson distribution (average rate of events)
lambda_param = 3

# Define a range of values for which to calculate the CDF
values_to_calculate_cdf = np.arange(0, 10) # Calculate CDF for values from 0 to 9

print(f"Calculating CDF for a Poisson distribution with lambda = {lambda_param}")

# Calculate the CDF for each value in the range
cdf_values = poisson.cdf(values_to_calculate_cdf, mu=lambda_param)

# Print the values and their corresponding CDF
print("\nValue | CDF")
print("------|------")
for value, cdf in zip(values_to_calculate_cdf, cdf_values):
  print(f"{value:<5} | {cdf:.4f}")

# Interpretation of the CDF:
print("\nInterpretation of the CDF:")
print(f"- The CDF at a specific value 'x' represents the probability of observing 'x' or fewer events in the given interval.")
print(f"- For example, the CDF at {values_to_calculate_cdf[2]} ({cdf_values[2]:.4f}) means there is a {cdf_values[2]:.2%} probability of observing {values_to_calculate_cdf[2]} or fewer events.")
print(f"- As the value of x increases, the CDF approaches 1, indicating that it becomes increasingly certain to observe a number of events less than or equal to that value.")
```

28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
  """
  Calculates the skewness and kurtosis of a dataset.

  Args:
    data: A NumPy array of numerical data.

  Returns:
    A tuple containing the skewness and kurtosis of the dataset.
  """
  # Calculate skewness
  data_skewness = skew(data)

  # Calculate kurtosis
  data_kurtosis = kurtosis(data) # By default, this is the excess kurtosis (kurtosis - 3)

  return data_skewness, data_kurtosis

# Generate a dataset from a normal distribution
mean = 0
std_dev = 1
sample_size = 1000
normal_data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

# Calculate skewness and kurtosis for the normal dataset
skewness_normal, kurtosis_normal = calculate_skewness_kurtosis(normal_data)

print(f"Skewness of the normal dataset: {skewness_normal:.4f}")
print(f"Kurtosis of the normal dataset: {kurtosis_normal:.4f}")

# Example with a non-normal dataset (e.g., exponential)
exponential_data = np.random.exponential(scale=1, size=1000)
skewness_exp, kurtosis_exp = calculate_skewness_kurtosis(exponential_data)

print(f"\nSkewness of the exponential dataset: {skewness_exp:.4f}")
print(f"Kurtosis of the exponential dataset: {kurtosis_exp:.4f}")
```
