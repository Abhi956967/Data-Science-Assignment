4.  How does the average discount percentage vary across categories

```python
import pandas as pd

# Load dataset
df = pd.read_csv("amazon.csv")

# Drop missing values
df = df.dropna(subset=['actual_price', 'discounted_price', 'category'])

# Clean and convert prices
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)

# Avoid division by zero
df = df[df['actual_price'] > 0]

# Calculate discount percentage
df['discount_percent'] = ((df['actual_price'] - df['discounted_price']) / df['actual_price']) * 100

# Group by category and compute average discount
avg_discount_by_category = df.groupby('category')['discount_percent'].mean().sort_values(ascending=False)

# Display result
print("Average discount percentage by category:\n")
print(avg_discount_by_category)
```

7. What is the correlation between discounted_price and rating

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("amazon.csv")

# Drop rows with missing values in required columns
df = df.dropna(subset=['discounted_price', 'rating'])

# Clean discounted_price column (remove ₹ symbol and commas)
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)

# Convert rating to numeric (in case of invalid values)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop rows with invalid ratings
df = df.dropna(subset=['rating'])

# Compute correlation
correlation = df[['discounted_price', 'rating']].corr()
print(correlation)
```

8.  What are the Top 5 categories based on the highest rating

```python
import pandas as pd

# Load dataset
df = pd.read_csv("amazon.csv")

# Drop missing values in 'category' and 'rating'
df = df.dropna(subset=['category', 'rating'])

# Convert rating to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop rows where rating could not be converted
df = df.dropna(subset=['rating'])

# Calculate average rating per category
avg_rating_by_category = df.groupby('category')['rating'].mean()

# Sort and get top 5 categories
top_5_categories = avg_rating_by_category.sort_values(ascending=False).head(5)

# Display the result
print("Top 5 categories by highest average rating:\n")
print(top_5_categories)
```

4. Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using
a countplot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("spotify.csv")

# Drop missing and duplicate values
df = df.dropna(subset=['artists']).drop_duplicates()

# Count tracks per artist
top_artists = df['artists'].value_counts().head(10)  # Show only top 10 for clarity

# Set plot style
sns.set(style="whitegrid")

# Plot countplot
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['artists'].isin(top_artists.index)],
              y='artists',
              order=top_artists.index,
              palette='viridis')
plt.title('Top 10 Artists by Number of Tracks')
plt.xlabel('Track Count')
plt.ylabel('Artist')
plt.tight_layout()
plt.show()

# Display top artist
top_artist = top_artists.idxmax()
top_count = top_artists.max()
print(f"\n🎤 Artist with the highest number of tracks: {top_artist} ({top_count} tracks)")
```

Calculate the average popularity for each of the top 5 artists in the filtered dataset.

## Filter data for top artists

### Subtask:
Create a subset of the DataFrame containing only tracks by these top 5 artists.

Filter the DataFrame to include only the tracks by the top 5 artists.



```python
# Filter the DataFrame to include only tracks by the top 5 artists
top_artists_df = df[df['Artist'].isin(top_5_artist_names)].copy()

# Display the first few rows of the filtered DataFrame to verify
display(top_artists_df.head())
```

## Find most popular track per artist

### Subtask:
For each of the top 5 artists, find the track with the highest popularity.


**Reasoning**:
For each of the top 5 artists, find the track with the highest popularity and store the artist and track name.



```python
most_popular_tracks_by_artist = {}

for artist in top_5_artist_names:
    # Filter tracks for the current artist
    artist_tracks = top_artists_df[top_artists_df['Artist'] == artist]

    # Find the index of the track with the highest popularity for this artist
    most_popular_track_index = artist_tracks['Popularity'].idxmax()

    # Get the track name using the index
    most_popular_track_name = artist_tracks.loc[most_popular_track_index]['Track Name']

    # Store the result
    most_popular_tracks_by_artist[artist] = most_popular_track_name

# Display the result
print("Most popular track for each of the top 5 artists:")
for artist, track_name in most_popular_tracks_by_artist.items():
    print(f"{artist}: {track_name}")
```

List the name of the most popular track for each of the top 5 artists.

Print the most popular track name for each artist stored in the most_popular_tracks_by_artist dictionary.



```python
# Print the most popular track name for each artist
print("Most popular track for each of the top 5 artists:")
for artist, track_name in most_popular_tracks_by_artist.items():
    print(f"{artist}: {track_name}")
```
